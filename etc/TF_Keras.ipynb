## Tensorflow architecture
  * [Layer-5] Tensorflow Estimators <- High-level, Object-Oriented API
  * [Layer-4] tf.layers, tf.losses, tf.metrics <- Reusable libraries for common model components
  * [Layer-3] Python Tensorflow <- Provides Ops, which wrap C++ kernels
  * [Layer-2] C++ Tensorflow
  * [Layer-1] CPU, GPU, TPU <- Kernels work on one or more platforms

### [즉시 실행 모드(Eager Mode) 지원]
  * Tensorflow를 Python에서처럼 바로 실행함
  * 1.x 버전에서는 '그래프'를 생성하고 초기화한 후에 세션을 통해 **값을 흐르게 하는 작업**을 진행해야 함

```python
a=tf.constant(3)
b=tf.constant(2)
print(tf.add(a, b))
print(a+b)
```

```python
print(tf.subtract(a, b))
print(a-b)
```

```python
print(tf.multiply(a, b))
print(a*b)
```

```python
print(tf.divide(a, b))
print(a/b)
```

### [Tensorflow ↔ Numpy]
  * numpy()
  * tf.convert_to_tensor()

```python
c=tf.add(a, b).numpy() #계산 결과를 numpy()로 변환함
print(type(c))
```

```python
c_square=np.square(c, dtype=np.float32)
c_tensor=tf.convert_to_tensor(c_square)
print(c_tensor)
print(type(c_tensor))
```

### [Numpy처럼 사용하기]

```python
print(t[:, 1:]) #Slicing
```

```python
print(t)
t[..., 1, tf.newaxis] #...을 Ellipsis라고 하며, 이는 앞의 모든 차원을 유지하라는 의미. 마지막 차원에서 1번(인덱스)을 선택하고 그 자리에 새로운 축을 추가함(선택된 1열의 값은 새로운 행이 됨)
```

```python
t+10 #모든 요소에 10을 더하게 됨
```

### [Type 변환]
   * Tensor의 기본 dtype
      - float형 Tensor: float32
      - int형 Tensor: int32
   * 연산 시 Tensor의 type이 반드시 일치해야 함
      - float32 ↔ float32
      - int32 ↔ int32
      - float32 ↔ int32 (x)
   * type 변환은 tf.cast()를 사용함

```python
a=tf.constant(2)
print(a)
b=tf.constant(2.0)
print(b)
#c=a+b #type이 서로 맞지 않기 때문에 에러 발생
print("a_dtype:", type(a), "b_dtype:", type(b))
c=tf.cast(a, dtype=tf.float32)+b
print(c)
```
