```python
from collections import deque
in_data = [1, 2, 3, 4, 5]
stack = deque()
for d in in_data:
    stack.append(d)
print(stack)
for _ in range(len(stack)):
    d = stack.pop()
    print(d)
else:
    print("Stack Test finished!")
```

```python
in_data = [1, 2, 3, 4, 5]
stack = deque()
for d in in_data:
    stack.append(d)
print(stack)
for _ in range(len(stack)):
    d = stack.popleft()
    print(d)
else:
    print("Queue Test finished!")
```

```python
def vector_addition(*args):
    return [sum(t) for t in zip(*args)] #args를 unpacking함

u = [2, 2, 3, 5, 7]
v = [2, 3, 6, 8, 3]
z = [3, 5, 1, 1, 1]
row_vectors = [[2, 2, 3, 5, 7], [2, 3, 6, 8, 3], [3, 5, 1, 1, 1]]
print(vector_addition(u, v, z))
print(vector_addition(*row_vectors)) #unpacking하여 전달해야 함
```

```python
from enum import Enum
class Color(Enum):
    RED = 'red'
    GREEN = 'green'
    BLUE = 'blue'

color = Color(input("Enter your choice of 'red', 'blue' or 'green': "))

match color:
    case Color.RED:
        print("I see red!")
    case Color.GREEN:
        print("Grass is green")
    case Color.BLUE:
        print("I'm feeling the blues :(")

print(color)
print(color.name)
```
