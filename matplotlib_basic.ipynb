```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

mpl.__version__
```

```python
plt.style.use(['seaborn-v0_8-notebook', 'ggplot'])
```

```python
fig=plt.figure()
ax=plt.axes()
#plt.plot([0,0.2,0.4,0.6,0.8,1]*5);
x=np.arange(0,10,0.1)
plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

```python
plt.plot(np.random.randn(50).cumsum(), 'y-.');
plt.plot(np.random.randn(50).cumsum(), linestyle='dashed', color='k');
plt.plot(np.random.randn(50).cumsum(), linestyle='dashed', color='#123456');
```

```python
plt.plot(np.random.randn(50));
plt.xlim(-1,50);
plt.ylim(-2,2);
plt.axis('tight');
plt.axis([-1,50,-2,2]);
plt.axis('equal');
```

```python
plt.plot(np.random.randn(100), label="A");
plt.plot(np.random.randn(100), label="B");
plt.legend(loc='best');
plt.title('Random Plot')
plt.xlabel('x')
plt.ylabel('random value')
```

```python
set([f.name for f in mpl.font_manager.fontManager.ttflist])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'Arial','color':'darkred','size':15, 'weight': 'bold'}
font3 = {'family':'Bahnschrift','color':'darkred','size':15}
plt.plot(np.random.randn(100), label="A");
plt.plot(np.random.randn(100), label="B");
plt.legend(loc='best');
plt.title('Random Plot', fontdict=font1)
plt.xlabel('x', fontdict=font2)
plt.ylabel('random value', fontdict=font3)
```

```python
fig, ax = plt.subplots()
ax.plot(np.random.randn(100), color='blue', label="A");
ax.plot(np.random.randn(100), color='red', label="B");
ax.plot(np.random.randn(100), color='green', label="C");
ax.legend(loc='upper right', frameon=True, framealpha=1, shadow=True, borderpad=1, ncol=2);
ax.set_title('Random Plot', fontdict=font1)
ax.set_xlabel('x', fontdict=font2)
ax.set_ylabel('random value', fontdict=font3)
ax.legend(fancybox=True, shadow=True, borderpad=1, framealpha=1, ncol=1)
```
